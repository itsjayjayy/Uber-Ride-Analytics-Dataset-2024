```python
# Filter only completed rides
completed_rides = df[df['Booking Status'] == 'Completed']

# Count rides per vehicle type
vehicle_counts = completed_rides['Vehicle Type'].value_counts()

print("Completed rides by Vehicle Type:")
print(vehicle_counts)

# Vehicle type with the most completed rides
top_vehicle = vehicle_counts.idxmax()
top_count = vehicle_counts.max()

print(f"\nðŸš– Vehicle type with most completed rides: {top_vehicle} ({top_count} rides)")
```

```python
# Convert to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Peak hours
df['Time'].dt.hour.value_counts().sort_index().plot(kind='bar')
plt.title("Bookings by Hour of Day")
plt.show()

```
